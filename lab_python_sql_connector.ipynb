```python
from sqlalchemy import text
import pandas as pd

def rentals_month(engine, month, year):
    with engine.connect() as connection:
        query = text("SELECT * FROM rental WHERE YEAR(rental_date) = :year AND MONTH(rental_date) = :month")
        result = connection.execute(query, {"year": year, "month": month})
        df = pd.DataFrame(result.fetchall(), columns=result.keys())
    return df

result = rentals_month(engine, 5, 2005)
result
```

```python
# task 3

def rental_count_month(df, month, year):
    result2 = pd.DataFrame(df.groupby(["customer_id"])["rental_id"].count())
    result2 = result2.rename(columns={"rental_id": f"rentals_{month}_{year}"})
    return result2

df1_=rental_count_month(result, "05", 2005)
df1_
```

```python
result3 = rentals_month(engine, 6, 2005)  
result3
```

```python
df2_ = rental_count_month(result3, "06", 2005)
df2_
```

```python
# task 4

def compare_rentals(df1, df2):
    new = df1.merge(df2, how="outer", on="customer_id")
    new.fillna(0, inplace = True)
    new["difference"] = new.iloc[:, 1] - new.iloc[:, 0]
    
    return new

# Example usage
compare_rentals(df1_, df2_)
```
